### Problem J2: Silent Auction

```python
N = int(input())
bid_record = 0
name_record = ""
for i in range(N):
  name = input()
  bid = int(input())
  if bid > bid_record:
    bid_record = bid
    name_record = name
print(name_record)

   
```

### Problem J5/S2: Modern Art

```python
M = int(input())
N = int(input())
K = int(input())
table = []
for _ in range(M):
  line = []
  for j in range(N):
    line.append(0)
  table.append(line)

for i in range(K):
  s = input()
  lst = s.split()
  if lst[0] == "R":
    for i in range(N):
      table[int(lst[1]) - 1][i] += 1
  else:
    for row in table:
      row[int(lst[1]) - 1] += 1

count = 0
for i in range(M):
  for j in range(N):
    if table[i][j]% 2 == 1:
      count += 1

print(table)
print(count)
```
