### Problem J4: Cyclic Shifts

```python
T = input()
S = input()

def cyclic_shift(T, S):
  for i in range(len(S)):
    s = S[i:] + S[0:i]
    if s in T:
      return "yes"
  return "no"
print(cyclic_shift(T, S))
```

### Problem J5/S2: Escape Room

```python
N = int(input())
M = int(input())

lst = []
for i in range(N):
  s = input()
  lst.append(s.split())
```

```python
def factor(value):# it returns the real index of the jump-to cube in the lst
  global M
  global N
  result = []
  for factor1 in range(1, int(value**0.5) + 1): ### remember!!!!
    if value % factor1 == 0:
      factor2 = value // factor1
      if factor1 <= M and factor2 <= N:
        result.append([factor1 - 1, factor2 - 1])
      if factor1 <= N and factor2 <= M:
        result.append([factor2 - 1, factor1 - 1])
  return result
```

```python
M = int(input())
N = int(input())
table = []
graph = {}
for i in range(M):
  s = input()
  #list comprehension - when you want to run a for loop and store the results in a list
  #syntax: newList = [ expression(element) for element in oldList if condition ]
  line = []
  for node in s.split():
    value = int(node)
    line.append(value)
    graph[value] = factor(value) # construct the graph
  table.append(line)
```

```python
def bfs(root, end):
  visited = []
  queue = []
  queue.append(root)
  while queue:
    head = queue.pop(0)
    print(head, end = "->")
    if head == end:
      print('yes')
      return
    for child in graph[head]:
      if child not in visited:
        visited.append(child)
        queue.append(child)
  print('no')
bfs(table[1][1],table[M-1][N-1])
```
