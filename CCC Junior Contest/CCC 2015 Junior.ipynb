### Problem J1: Special Day

```python
m = int(input())
d = int(input())

if m < 2 and d < 18:
  print("Before")
elif m > 2 and d > 18:
  print("After")
else:
  print("Special")
```

### Problem J2: Happy or Sad

```python
s = input()
happy_count = s.count(":-)")
sad_count = s.count(":-(")

if happy_count == 0 and sad_count == 0:
  print("none")
elif happy_count > sad_count:
  print("happy")
elif happy_count < sad_count:
  print("sad")
else:
  print("unsure")
```
