# Matplotlib
Matplotlib is widely used for data visualization like for plotting histograms, line plots, and heat plots.

### **Pylot**

Most of the Matplotlib utilities lies under the Pylot submodule. When we import that, we usually use the alias plt.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0,6])
y = np.array([0,100])

plt.plot(x,y)
plt.show()
```

**The plot() function** is used to draw points (markers) in a diagram.

By default, the plot() fuction draws a line from point to point.

Pramater 1 - an array, points on the x-axis.

Pramater 2 - an array, points on the y-axis.

**Exercise:**

Plot a line from (1,3) to (8,10).

```python
plt.plot(np.array([1, 3]), np.array([8, 10]))
plt.show()
```

### **Markers**

You can use the keyword argument "marker" to emphasize each point with a specific marker.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])

plt.plot(y, marker = 'o') # '*' - star, '.' - point, ',' - pixel, 'x' - X
plt.show()
```

Format string - fmt

Shortcut string notation - specify the marker.

This parameter is also called fmt.

Syntax: marker|line|color, e.g: 'o:r'

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])

plt.plot(y, 'o-m') # '*' - star, '.' - point, ',' - pixel, 'x' - X
plt.show()
```

Marker Reference:

Marker: 
*   '*' - star
*   '.' - point
*   ',' - pixel
*   'x' - X

Line: 
*   '-' - solid line
*   ':' - Dotted line
*   '--' - Dashed line
*   '-.' - Dashed/dotted line

Color:
*   'r' - red
*   'b' - blue
*   'g' - green
*   'y' - yellow
*   'k' - black
*   'w' - white
*   'c' - Cyan
*   'm' - Magenta



**Marker size**

keyword is markersize or ms

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])

plt.plot(y, marker = 'o', ms = 20)
plt.show()
```

### **Line**


keyword is **linestyle**, or ls, to change the style of the plotted line.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,1,3])

plt.plot(y, ls = 'dashdot')
plt.show()
```

Linestyle reference:

* dotted or ':'
* dashed or '--'
* solid or '-' (by default)
* dashdot or '-.'
* None or '' or ' '

**Linecolor**: keyword is color or c

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,1,3])

plt.plot(y, ls = 'dashdot', c = 'hotpink') # OR 'g" or {color name}
plt.show()
```

**Line Width**: keyword is linewidth or lw

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([2,4,1,3])

plt.plot(y, ls = 'dashdot', c = 'hotpink', lw = '6.2') # lw is a 'float'
plt.show()
```

**Multiple lines**

Example1: only y-axis values are provided

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([2,4,1,3])
y2 = np.array([3,8,1,10])

plt.plot(y1)
plt.plot(y2)
plt.show()
```

Example2: both x and y values are provided.

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0,1,2,3])
y1 = np.array([2,4,1,3])
x2 = np.array([0,1,2,3])
y2 = np.array([3,8,1,10])

plt.plot(x1,y1,x2,y2)
plt.show()
```

### **Labels and Title**

Use **xlabel()** to set x-axis label. Use **ylabel()** to set y-axis label.

Use **title()** to set the title.

Use **fontdict** parameter to set font styles.

Use **loc** parameter to set the position of the title. (center(by default), left, right)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([10,15,20,25,30,35,40,45,50])
y = np.array([200,210,220,230,240,250,260,270,280])

title_font = {'family':'serif', 'color': 'blue', 'size': 20}
label_font = {'family':'serif', 'color': 'pink', 'size': 16}

plt.plot(x,y)
plt.title("Something fun", fontdict = title_font, loc = "left")
plt.xlabel("Days", fontdict = label_font)
plt.ylabel("Price", fontdict = label_font)
plt.show()
```

### **Subplot**

With subplot, we can draw mutiple plots in one figure.

```python
import matplotlib.pyplot as plt
import numpy as np

# plot1 
x1 = np.array([0,1,2,3])
y1 = np.array([2,4,2,6])
plt.subplot(1,2,1) # this figure has 1 row, 2 columns, and the subplot is in the first plot
plt.plot(x1,y1)
plt.title("plot1")

# plot2 
x2 = np.array([0,1,2,3])
y2 = np.array([40,30,60,30])
plt.subplot(1,2,2) # this figure has 1 row, 2 columns, and the subplot is in the second plot
plt.plot(x2,y2)
plt.title("plot2")

plt.suptitle("Two plots")
plt.show()
```

### 2019 J2 - Time to Decompress

```python
decoded_message = []
num = int(input())
for i in range(num):
  chars = input().split()
  n = int(chars[0]) # occurence
  char = chars[1] # character
  decoded_message.append(char * n) # Key-learning: char*N- stirng operation
for line in decoded_message:
  print(line)
```

### 2019 J3 - Decoded version of cold compress

```python
N = int(input())
lines = []

for i in range(N):
  line = input()
  lines.append(line)

# 6 7 6 . 12 T
for line in lines:
  lst = line.split()
  result = "" 
  for j in range(0,len(lst),2): # Key-learning: use step in range to handle a list in steps
    num = int(lst[j])
    char = lst[j+1]
    result += char*num
  print(result)
```
