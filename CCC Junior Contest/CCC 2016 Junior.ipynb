### Problem J1: Tournament Selection

```python
lst = []
for i in range(6):
  s = input()
  lst.append(s)
win = lst.count("W")
if win >= 5 and win <= 6:
  print(1)
elif win >= 3 and win <= 4:
  print(2)
elif win >= 1 and win <= 2:
  print(1)
else:
  print(-1)
```

### Problem J2: Magic Squares

```python
lst = []
row_sum = []
for _ in range(4):
  s = input()
  row = s.split()
  sum = 0
  for num in row:
    sum += int(num)
  row_sum.append(sum)
  lst.append(s.split())

first_sum = row_sum[0]

condition = True
for sum in row_sum:
  if first_sum != sum:
    condition = False

for i in range(4):
  sum = 0
  for sub_lst in lst:
    sum += int(sub_lst[i])
  if first_sum != sum:
    condition = False

if condition == True:
  print("magic")
else:
  print("not magic")
```
