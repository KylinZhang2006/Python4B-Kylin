Input: 'HelloKylin'

Output: 'NILYKOLLEH'

```python
# Your code
s = 'HelloKylin'
print("".join(list(reversed(list(s.upper())))))

```

Create a Student ~~list~~ class with an age field. Use lambda function to return the younger Student.

Input (You decide the format): StudentA(age: 15), StudentB(age: 12)

Output (You decide the format): StudentB

```python
# Your code
StudentA = 15
StudentB = 12

Max = lambda a, b : a if(a > b) else b 
print("Oldest Student has an age of:", Max(StudentA, StudentB))
```

```python
# Student Class
class Student:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def get_name(self):
    return self.name
  def get_age(self):
    return self.age

studentA = Student("A", 15)
studentB = Student("B", 12)

older = lambda a, b : a if(a.get_age() > b.get_age()) else b
print("The oldest student is:", older(studentA, studentB).get_name())

```

✅*score: 10/10*

Solution:

Typo, it should be Student class instead of Student list. 

```python
class Student:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def get_age(self):
    return self.age
  def get_name(self):
    return self.name
StudentA = Student("A", 15)
StudentB = Student("B", 12)
Max = lambda a, b : a if(a.get_age() > b.get_age()) else b 
older_student = Max(StudentA, StudentB)
print("Older: Student{} has an age of {}.".format(older_student.get_name(), older_student.get_age()))
```
