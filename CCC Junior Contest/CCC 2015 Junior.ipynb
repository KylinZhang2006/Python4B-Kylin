### Problem J2: Happy or Sad

```python
s = input()
happy_count = s.count(":-)")
sad_count = s.count(":-(")

if happy_count == 0 and sad_count == 0:
  print("none")
elif happy_count > sad_count:
  print("happy")
elif happy_count < sad_count:
  print("sad")
else:
  print("unsure")
```

### Problem J3: Rovarspr Â¨ aket

```python
alphabet_list = ["a", "b", "c", "d", "e", "f", "g",
                 "h", "i", "j", "k", "l", "m", "n", 
                 "o", "p", "q", "r", "s", "t", 
                 "u", "v", "w", "x", "y", "z"]
vowel = ["a", "e", "i", "o", "u"]
vowel_index = [0, 4, 8, 15, 20]
s = input()
result = ""

for char in s:
  result += char
  if char in vowel:
    continue
  else:
    index = alphabet_list.index(char)
    closest_vowel_distance = 27
    closest_vowel = ""
    for v in vowel_index:
      if abs(index - v) < closest_vowel_distance:
        closest_vowel_distance = abs(index - v)
        index_vowel = vowel_index.index(v)
        closest_vowel = vowel[index_vowel]
    result += closest_vowel
    if alphabet_list[index + 1] not in vowel:
      result += alphabet_list[index + 1]
    else:
      result += alphabet_list[index + 2]

print(result)
```

### Problem J4: Wait Time

```python
M = int(input())

friend_to_time = {} # {friend: [R_time, wait_time]}
time = -1
for i in range(M):
  line = input().split()
  letter = line[0]
  friend = int(line[1])
  
  if letter == "W": # friend is waiting time
    time += friend - 1
  
  else:
    time += 1
  
  if letter != "W":
    if letter == "R":
      if friend in friend_to_time:
        friend_to_time[friend][0] = time
      else:
        friend_to_time[friend] = [time, 0]
    elif letter == "S":
      friend_to_time[friend][1] = friend_to_time[friend][1] + (time - friend_to_time[friend][0])
      friend_to_time[friend][0] = -1 
  
```

```python
friends = list(friend_to_time.keys())
friends.sort()
for friend in friends:
  if friend_to_time[friend][0] == -1:
    print("{} {}".format(friend, friend_to_time[friend][1]))
  else:
    print("{} {}".format(friend, -1))
```
