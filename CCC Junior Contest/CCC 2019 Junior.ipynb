### Problem J1: Winning Score

```python
A = []
count = 3
for i in range(3):
  s = input()
  A.append(int(s)*count)
  count -= 1

B = []
count = 3
for j in range(3):
  s = input()
  B.append(int(s)*count)
  count -= 1

if sum(A) > sum(B):
  print("A")
elif sum(A) == sum(B):
  print("T")
else:
  print("B")
```

### Problem J2: Time to Decompress

```python
lines = []
num = int(input())
for i in range(num):
  s = input()
  chars = s.split()
  result = ""
  for j in range(int(chars[0])):
    result += chars[1]
  lines.append(result)
for line in lines:
  print(line)
```
