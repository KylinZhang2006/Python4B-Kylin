```python
"""Question 1"""
import matplotlib.pyplot as plt
import numpy as np
# Plot 1
x1 = np.arange(100)
y1 = np.random.randint(20, 30, 100)
plt.subplot(2, 2, 1)
plt.xlabel("x1-value")
plt.ylabel("y1-value")
plt.title("plot1")
plt.plot(x1, y1)

# Plot 2
x2 = np.arange(100)
y2 = np.ones(100, dtype = int)
plt.subplot(2, 2, 2)
plt.xlabel("x2-value")
plt.ylabel("y2-value")
plt.title("plot2")
plt.plot(x2, y2)

# Plot 3
x3 = np.arange(100)
y3 = x3 + 10*np.random.randn(100)
plt.subplot(2, 2, 3)
plt.xlabel("x3-value")
plt.ylabel("y3-value")
plt.title("plot3")
plt.plot(x3, y3)

plt.show()
```
